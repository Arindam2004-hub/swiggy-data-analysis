```python
file1['actual_dilevery_time']=file1['delivery_time']-file1['order_time']
file1['expected_dilevery_time.']=file1['expected_delivery_time']-file1['order_time']
file1
```

```python
# calculate and column price x quantity
file1['net_price'] = file1['price'] * file1['quantity']
file1
```

```python
import pandas as pd

# Step 1: Group by city and order_status to count orders
grouped = file1.groupby(['city_x', 'order_status']).size().reset_index(name='order_count')

# Step 2: Calculate total orders per city
total_orders = grouped.groupby('city_x')['order_count'].transform('sum')

# Step 3: Calculate percentage for each status
grouped['percentage'] = (grouped['order_count'] / total_orders) * 100

# Step 4: Reshape the data for better readability (optional)
result = grouped.pivot(index='city_x', columns='order_status', values='percentage').fillna(0)

print(result)

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by 'City' and 'dish_ordered' and sum the 'Orders'
grouped = file2.groupby(['city_x', 'dish_ordered'])['net_price'].sum().reset_index()

# Find the item with the maximum orders in each city
max_order = grouped.loc[grouped.groupby('city_x')['net_price'].idxmax()]

# Plotting the data
plt.figure(figsize=(15, 10))

# Plot the price data by city
plt.bar(max_order['city_x'], max_order['dish_ordered'])

# Adding labels and title
plt.xlabel('City')
plt.ylabel('Dish name')
plt.title('Most ordered dish in each city')

# Show the plot
plt.show()

```

```python
# Group by 'City' and 'dish' and sum the 'price'
grouped = file2.groupby(['city_x','dish_ordered'])['net_price'].sum().reset_index()

# Find the item with the maximum orders in each city
max_orders = grouped.loc[grouped.groupby('city_x')['net_price'].idxmax()]

print(max_orders)
```

```python
# Group by city and restaurant, then aggregate net price and reviews
grouped = file2.groupby(['city_x', 'restaurant_name']).agg(
    total_net_price=('net_price', 'sum'),
    total_reviews=('total_reviews', 'sum')
).reset_index()

# Sort within each city by total_net_price in descending order
grouped = grouped.sort_values(['city_x', 'total_net_price'], ascending=[True, False])

# Get top 5 restaurants for each city
top_restaurants = grouped.groupby('city_x').head(5)
print(top_restaurants)
```
